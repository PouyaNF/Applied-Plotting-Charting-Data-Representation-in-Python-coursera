```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import matplotlib.ticker as ticker
import pandas as pd
import numpy as np
import mplleaflet    #Convert Matplotlib plots into Leaflet web maps


df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')

# Convert all df to degrees Celsius:
df['Data_Value'] = df['Data_Value']*0.1

# separation of Year and Month/day into 2 different dfs
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month_Day'] = df['Date'].dt.strftime('%m-%d')

# Remove February 29 from df 
df = df[df['Month_Day']!='02-29']


df.head(5)
```

```python
# choosing years between 2005 to 2014 ,  then groupin data by each day, then picking max and min value fr temp (Data_Value)
max_temp = df[(df.Year >= 2005) & (df.Year < 2015) & (df['Element'] == 'TMAX')].groupby(['Month_Day'])['Data_Value'].max()
min_temp = df[(df.Year >= 2005) & (df.Year < 2015) & (df['Element'] == 'TMIN')].groupby(['Month_Day'])['Data_Value'].min()


max_temp.head()
```

```python
# Record breaking df:
# adding min and max values for each day of year for  data between 2005-2014 
df = df.merge(max_temp.reset_index(drop=False).rename(columns={'Data_Value':'Max_temp'}), on='Month_Day', how='left')
df = df.merge(min_temp.reset_index(drop=False).rename(columns={'Data_Value':'Min_temp'}), on='Month_Day', how='left')

# Creating 2 dfs for record breaking data for days of 2015 
record_high = df[(df.Year==2015) & (df.Data_Value > df.Max_temp)]
record_low  = df[(df.Year==2015) & (df.Data_Value < df.Min_temp)]

record_high.head()
```

```python
# Plotting 

date_index = np.arange('2015-01-01','2016-01-01', dtype='datetime64[D]')
plt.figure(figsize = (15,15))

plt.plot(date_index,max_temp,color='#E77613', linewidth=1) # Max in red.
plt.plot(date_index,min_temp,color='#10C2ED', linewidth=1) # Min in blue.
plt.scatter(record_high.Date.values, record_high.Data_Value.values, color='#B92B2B', s=7)
plt.scatter(record_low.Date.values, record_low.Data_Value.values, color='#083FFA', s=7)

# Set x and y limits.
ax = plt.gca()
ax.axis(['2015/01/01','2015/12/31',-50,50])


# Set axis names and title:
plt.xlabel('Time of Year', fontsize=15)
plt.ylabel('Temperatures (C) ', fontsize=15)
plt.title('Ranges of Temperatures for Ann Arbour, Michigan (2005-2014) and Record-Breaking Temperatures in 2015 ', fontsize=15)

# Create legend and title
# loc=0 provides the best position for the legend
plt.legend(['Highest Temperatures (2005-2014)','Lowest Temperatures (2005-2014)','Record breaking high in 2015','Record breaking low in 2015'],
           loc=0,frameon=False,fontsize = 15 )


# Fill colour between highs and lows:
# alpha adjusts darkness of the shade.
ax.fill_between(date_index, max_temp, min_temp, facecolor='#C3C6B8', alpha=0.25)

# Where you locate the major and minor ticks:
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_minor_locator(dates.MonthLocator(bymonthday=15)) # Put the label at the minor tick so it's in the center.
#ax.yaxis.set_minor_locator()

# What you put at the ticks:
ax.xaxis.set_major_formatter(ticker.NullFormatter())
ax.xaxis.set_minor_formatter(dates.DateFormatter('%b'))

# 1 refers to the bottom of the plot for xticks and the left for yticks
# 2 refers to the top of the plot for xticks and the right for yticks
for tick in ax.xaxis.get_minor_ticks():
    tick.tick1line.set_markersize(0) # Make small ticker disappear
    #tick.label1.set_horizontalalignment('center')

plt.show()
```
